# Finding URL

```python
searchString='iphone'
url = 'https://www.flipkart.com/search?q={}&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'.format(searchString)
```

```python
data=requests.get(url)
```

```python
Text=BeautifulSoup(data.content,'html.parser')
```

```python
products=Text.find_all('div',{'class':'_3FqKqJ E2-pcE _1q8tSL _2pi5LC col-12-12'})
products
```

```python
products_1=Text.find_all('div',{'class':'_3FqKqJ'})
products_2=[product.find('div',{'class':'E2-pcE'}) for product in products_1]
products_3=[product.find('div',{'class':'_1q8tSL'}) for product in products_2]
products_4=[product.find('div',{'class':'_2pi5LC'}) for product in products_3]
products_5=[product.find_all('div',{'class':'col-12-12'}) for product in products_1]
dataLink=[]
# dataLink=products_5[0].find_all('a',{'class':'_2kHMtA'})
for product in products_5[0]:
    temp=product.find_all('div',{'class':'_2kHMtA'}) 
    if temp is not None and len(temp)!=0:
        dataLink.append(temp[0])
dataLink
```

```python
extractData.find('div',{'class':'_1YokD2 _3Mn1Gg col-9-12'}).find_all('div',{'class':'_1AtVbE col-12-12'})[1].find('div',{'class':'t-ZTKy'}).findchild
```

### Extracting Review Heading,Rating,Body

```python
extractData.find('div',{'class':'_1YokD2 _3Mn1Gg col-9-12'}).find_all('div',{'class':'_1AtVbE col-12-12'})[1].find('div',{'class':'t-ZTKy'}).findChild().findChild().get_text()
```

```python
temp=extractData.find('div',{'class':'_1YokD2 _3Mn1Gg col-9-12'}).find_all('div',{'class':'_1AtVbE col-12-12'})
print(len(temp))
for review in temp:
    if review.findChild('p') is None:
        print('*')
        continue
    print('Heading: '+review.findChild('p').get_text())
    print('Rating: '+review.findChild('p').find_previous_sibling().get_text())
    print('Body: '+review.find('div',{'class':'t-ZTKy'}).findChild().findChild().get_text())
    print('%%##'*35)
```
